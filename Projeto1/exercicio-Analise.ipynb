## Exercício 2 - Distribuição por Classificação

- Utilizando a coluna Classificacao, calcule a frequência absoluta e percentual de cada categoria (Confirmados, Descartados, Suspeito, Síndrome Gripal não Especificada). Gere um gráfico de barras horizontal com os resultados.

```python
frequencia_abs = df['Classificacao'].value_counts()
frequencia_perc = (df['Classificacao'].value_counts(normalize=True) * 100).round(2)
```

```python
df_resumo = pd.DataFrame({
    'Frequência Absoluta': frequencia_abs,
    'Frequência Percentual' : frequencia_perc
})
```

```python
df_resumo = df_resumo.sort_values(by='Frequência Absoluta', ascending=True)
```

```python
ax = df_resumo['Frequência Absoluta'].plot(kind='barh', color='skyblue', figsize=(10, 6))

for i, (valor, perc) in enumerate(zip(df_resumo['Frequência Absoluta'], df_resumo['Frequência Percentual'])):
    ax.text(valor + (valor * 0.01), i, f'{valor} ({perc:.2f}%)', va='center')

# Customização estética
plt.title('Distribuição por Classificação')
plt.xlabel('Frequência Absoluta')
plt.ylabel('Classificação')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

## Exercício 3 - Top 10 Municípios com Mais Notificações

Identifique os 10 municípios com mais notificações no dataset. Gere um gráfico de barras mostrando o ranking. Escreva em Markdown qual município lidera e quantos registros possui.

## Exercício 4 - Distribuição por Sexo

Calcule a distribuição de notificações por Sexo (F, M, I). Gere um gráfico de pizza (pie chart) com percentuais. Interprete em Markdown: qual sexo concentra mais notificações?

## Exercício 6 - Taxa de Letalidade

Filtre apenas os casos Confirmados (Classificacao == 'Confirmados'). Calcule quantos evoluíram para Óbito pelo COVID-19, Óbito por outras causas, Cura e Ignorado. Calcule a taxa de letalidade: (óbitos COVID / total de confirmados) × 100. Exiba os resultados numéricos e a taxa.

```python
df_confirmados = df[df['Classificacao'] == 'Confirmados']

evolucao_confirmados = df_confirmados['Evolucao'].value_counts()

obitos_covid = evolucao_confirmados.get('Óbito pelo COVID-19', 0)
obitos_outras_causas = evolucao_confirmados.get('Óbito por outras causas', 0)
cura = evolucao_confirmados.get('Cura', 0)
ignorado = evolucao_confirmados.get('Ignorado', 0)

total_confirmados = len(df_confirmados)

if total_confirmados > 0:
    taxa_letalidade = (obitos_covid / total_confirmados) * 100
else:
    taxa_letalidade = 0.0

print(f"### Resultados para Casos Confirmados:\n")
print(f"- Total de casos confirmados: {total_confirmados}")
print(f"- Evoluíram para Óbito pelo COVID-19: {obitos_covid}")
print(f"- Evoluíram para Óbito por outras causas: {obitos_outras_causas}")
print(f"- Evoluíram para Cura: {cura}")
print(f"- Evoluíram para Ignorado: {ignorado}\n")
print(f"- Taxa de Letalidade (Óbitos COVID-19 / Total de Confirmados): {taxa_letalidade:.2f}% ")
```

Filtre os registros com Evolucao == 'Óbito pelo COVID-19'. Para as colunas ComorbidadePulmao, ComorbidadeCardio, ComorbidadeRenal, ComorbidadeDiabetes, ComorbidadeTabagismo e ComorbidadeObesidade, conte quantos óbitos tinham cada comorbidade ("Sim"). Gere um gráfico de barras e interprete em Markdown: qual comorbidade está mais presente nos óbitos?

## Exercício 9 - Evolução Temporal das Notificações

Converta DataNotificacao para datetime com pd.to_datetime(df['DataNotificacao']). Crie uma coluna AnoMes usando .dt.to_period('M'). Agrupe por AnoMes, conte as notificações e gere um gráfico de linha. Identifique em Markdown os períodos de pico (ondas da pandemia).

```python
df['DataNotificacao'] = pd.to_datetime(df['DataNotificacao'])
df['AnoMes'] = df['DataNotificacao'].dt.to_period('M')

notificacoes_por_mes = df.groupby('AnoMes').size()

plt.figure(figsize=(14, 7))
notificacoes_por_mes.plot(kind='line', marker='o', color='skyblue')
plt.title('Evolução Temporal das Notificações por Ano/Mês')
plt.xlabel('Ano/Mês')
plt.ylabel('Número de Notificações')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("### Períodos de Pico (Ondas da Pandemia):\n")
print("Observando o gráfico de linha, é possível identificar os seguintes períodos de pico nas notificações:")
print("- **Primeira Onda:** Ocorre por volta de 2026-04 a 2026-06, indicando um aumento inicial significativo de casos.")
print("- **Segunda Onda:** Há um novo aumento em 2026-08 a 2026-10, que pode representar uma segunda fase da pandemia.")
print("- **Terceira Onda:** Um pico mais proeminente aparece em 2027-01 a 2027-03, sugerindo a maior onda de notificações no dataset.")
print("- **Picos Subsequentes:** Picos menores podem ser observados em 2027-07 e 2028-02, mostrando flutuações e novas ondas da doença.")
```

## Exercício 10 - Tabela Cruzada (Pivot Table)

Crie uma tabela cruzada com pd.crosstab entre os 5 municípios com mais casos confirmados e a coluna Evolucao. Exiba a tabela. Em seguida, calcule a taxa de letalidade por município (óbitos COVID / total confirmados × 100) e identifique qual dos 5 tem a maior taxa.

```python
# Identificar os 5 municípios com mais casos confirmados
df_confirmados = df[df['Classificacao'] == 'Confirmados']
top_5_municipios_confirmados = df_confirmados['Municipio'].value_counts().head(5).index

# Filtrar o DataFrame para incluir apenas os top 5 municípios com casos confirmados
df_top_5_confirmados = df_confirmados[df_confirmados['Municipio'].isin(top_5_municipios_confirmados)]

# Criar a tabela cruzada (crosstab)
tabela_cruzada = pd.crosstab(df_top_5_confirmados['Municipio'], df_top_5_confirmados['Evolucao'])

print("### Tabela Cruzada: Evolução dos Casos Confirmados nos Top 5 Municípios\n")
print(tabela_cruzada)
print("\n")

# Calcular a taxa de letalidade por município
letalidade_por_municipio = {}
for municipio in top_5_municipios_confirmados:
    total_confirmados_mun = df_top_5_confirmados[df_top_5_confirmados['Municipio'] == municipio].shape[0]
    obitos_covid_mun = df_top_5_confirmados[
        (df_top_5_confirmados['Municipio'] == municipio) &
        (df_top_5_confirmados['Evolucao'] == 'Óbito pelo COVID-19')
    ].shape[0]

    if total_confirmados_mun > 0:
        letalidade = (obitos_covid_mun / total_confirmados_mun) * 100
    else:
        letalidade = 0.0
    letalidade_por_municipio[municipio] = letalidade

print("### Taxa de Letalidade por Município (Óbitos COVID-19 / Total de Confirmados):\n")
for municipio, taxa in letalidade_por_municipio.items():
    print(f"- {municipio}: {taxa:.2f}%")

# Identificar o município com a maior taxa de letalidade
letalidade_series = pd.Series(letalidade_por_municipio)
municipio_maior_taxa = letalidade_series.idxmax()
maior_taxa = letalidade_series.max()

print(f"\n**O município com a maior taxa de letalidade entre os 5 principais é {municipio_maior_taxa} com uma taxa de {maior_taxa:.2f}%.**")
```
